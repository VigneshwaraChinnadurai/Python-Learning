```
import numpy as np
import matplotlib.pyplot as plt
import os
import math
import seaborn as sns
import pandas as pd
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize']=12,6
```

```
mov1=movieratings[movieratings.Genre.isin(['action','comedy','adventure','animation','drama'])]
mov2=mov1[mov1.Studio.isin(['Buena Vista Studios','Sony','Universal','WB','Paramount Pictuures','Fox'])]
```

```
print(mov2.head())
print(mov2.columns)
print(mov2.Genre.unique())
print(mov2.Studio.unique())
```

```
sns.set(style='darkgrid',palette='muted',color_codes=True)

v=sns.boxplot(data=mov2, x='Genre', y='Gross % US',orient='v',color='lightgray',showfliers=False)
plt.setp(v.artists,alpha=0.7)

sns.stripplot(data=mov2,x='Genre',y='Gross % US',hue='Studio',size=9,alpha=0.7)

v.axes.set_title('Domestic Gross % by Genre',fontsize=40)
v.set_xlabel('Genre',fontsize=30)
v.set_ylabel('Gross % US',fontsize=30)
v.legend(bbox_to_anchor=(1.05,1),loc=2,prop={'size':15},frameon=True,fancybox=True,shadow=True,)
plt.show()
```
